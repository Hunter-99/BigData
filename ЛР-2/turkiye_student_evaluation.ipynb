# Набор данных Turkiye Student Evaluation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

### Загрузка датасета

```python
df = pd.read_csv('source/turkiye-student-evaluation_generic.csv')
df.head()
```

### Статистические данные о датасете

```python
df.describe()
```

```python
df.info()
```

### Анализ главных компонента по предмету 3

```python
sns.countplot(df['class'])
```

#### Матрица коореляции

```python
corr = df[df['class'] == 3].corr()
plt.figure(figsize=(18,18))
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

#### Анализ главных компонент

```python
X = df[df['class'] == 3].iloc[:, 5:33]
pca = PCA(n_components=28, random_state=42)
X_pca = pca.fit_transform(X)
X_pca
```

#### Сколько информации мы сохранили из набора данных

```python
pca.explained_variance_ratio_.cumsum()[1]
```
